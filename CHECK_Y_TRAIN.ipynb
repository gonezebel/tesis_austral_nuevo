```python
import os
import pandas as pd

# Ruta a la carpeta que contiene los archivos y_train
INPUT_PATH = r"C:\Users\Gonzalo\Documents\GitHub\TesisAustral\Archivos_tesis\intermediate\selected"
OUTPUT_CSV = os.path.join(INPUT_PATH, "distribucion_clases_y_train.csv")

# Listado de archivos y_train
archivos_y = [f for f in os.listdir(INPUT_PATH) if f.startswith("y_train") and f.endswith(".parquet")]

distribuciones = []

for archivo in archivos_y:
    try:
        path = os.path.join(INPUT_PATH, archivo)
        df_y = pd.read_parquet(path).squeeze()
        if not isinstance(df_y, pd.Series):
            df_y = df_y.iloc[:, 0]  # En caso de venir como DataFrame de 1 columna

        counts = df_y.value_counts().sort_index()
        percentages = df_y.value_counts(normalize=True).sort_index() * 100

        for clase in counts.index:
            distribuciones.append({
                "archivo": archivo,
                "clase": clase,
                "cantidad": counts[clase],
                "porcentaje": round(percentages[clase], 2)
            })

    except Exception as e:
        print(f"❌ Error procesando {archivo}: {e}")

# Convertir a DataFrame
df_resultado = pd.DataFrame(distribuciones)

# Guardar a CSV
df_resultado.to_csv(OUTPUT_CSV, index=False)
print(f"✅ Distribución de clases guardada en: {OUTPUT_CSV}")
```
