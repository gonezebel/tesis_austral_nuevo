```python
import os
import pandas as pd

# Ruta donde están los modelos entrenados (ajustá si es necesario)
BASE_MODELS_PATH = r"C:\Users\Gonzalo\Documents\GitHub\TesisAustral\Archivos_tesis\models\XGBoost_01_Standard_2025-08-04"
OUTPUT_CSV = os.path.join(BASE_MODELS_PATH, "metricas_unificadas_por_clase.csv")

# Recolectar todos los archivos metricas_*.csv
metricas_archivos = []
for root, _, files in os.walk(BASE_MODELS_PATH):
    for file in files:
        if file.startswith("metricas_") and file.endswith(".csv"):
            metricas_archivos.append(os.path.join(root, file))

# Cargar y concatenar
df_unificado = pd.DataFrame()
for path in metricas_archivos:
    try:
        df = pd.read_csv(path)
        nombre_modelo = os.path.basename(path).replace("metricas_", "").replace(".csv", "")
        df["modelo"] = nombre_modelo
        df_unificado = pd.concat([df_unificado, df], ignore_index=True)
    except Exception as e:
        print(f"❌ Error al procesar {path}: {e}")

# Guardar CSV final
df_unificado.to_csv(OUTPUT_CSV, index=False)
print(f"\n✅ Archivo final guardado en: {OUTPUT_CSV}")
```
